### Benchmark

```python
df = pd.read_csv("../data/benchmark_naive_implementation.csv", index_col=0)
df.head()
```

```python
max(df["time_bf"]), max(df["time_cc"])
```

```python
x = df.groupby(["n"]).mean().index
y_bf = df.groupby(["n"]).mean().time_bf
y_cc = df.groupby(["n"]).mean().time_cc
plt.plot(x, y_bf, label="Brute force time")
plt.plot(x, y_cc, label="Color coding time")
plt.xlabel("Number of nodes")
plt.ylabel("Execution time(s)")
plt.title("Comparison of execution times by number of nodes in the graph")
plt.legend()
plt.show()
```

```python
x = df.groupby(["k"]).mean().index
y_bf = df.groupby(["k"]).mean().time_bf
y_cc = df.groupby(["k"]).mean().time_cc
plt.plot(x, y_bf, label="Brute force time")
plt.plot(x, y_cc, label="Color coding time")
plt.xlabel("Number of nodes")
plt.ylabel("Execution time(s)")
plt.title("Comparison of execution times by number of nodes in the tree")
plt.legend()
plt.show()
```

```python
x = df.groupby(["p"]).mean().index
y_bf = df.groupby(["p"]).mean().time_bf
y_cc = df.groupby(["p"]).mean().time_cc
plt.plot(x, y_bf, label="Brute force time")
plt.plot(x, y_cc, label="Color coding time")
plt.xlabel("Probability")
plt.ylabel("Execution time(s)")
plt.title("Comparison of execution times by probability for edge creation.")
plt.legend()
plt.show()
```
